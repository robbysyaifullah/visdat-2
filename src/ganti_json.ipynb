```python
import json
```

```python
with open('geojson_jabar.json') as fp:
    geojson_jabar = json.load(fp)
```

```python
geojson_jabar
```

```python
geojson_change = {}
geojson_change['type'] = 'FeaetureCollection'
geojson_change['features'] = []
for i in range(len(geojson_jabar['features'])):
    geojson_change['features'].append(geojson_jabar['features'][i])
    geojson_change['features'][i]['geometry']['type'] = 'Polygon'
    geojson_change['features'][i]['geometry']['coordinates'] = geojson_jabar['features'][i]['geometry']['coordinates'][0]
```

```python
geojson_change
```

```python
with open('geojson_change.json', 'w') as fp:
    json.dump(geojson_change, fp)
```
